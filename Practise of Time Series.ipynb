##### Bollinger Bands

```python
# Close 20 MA
amz["Close: 20 Day Mean"] = amz["Close"].rolling(20).mean()

# Upper = 20 MA + 2* std(20 MA)
amz["Upper"] = amz["Close: 20 Day Mean"]+ 2* (amz["Close"].rolling(20).std())

# Lower = 20 MA - 2* std(20 MA)
amz["Lower"] = amz["Close: 20 Day Mean"]- 2* (amz["Close"].rolling(20).std())

# Close
amz[["Close","Close: 20 Day Mean","Upper","Lower"]].plot(figsize = (16,12));
```

```python
amz[["Close","Close: 20 Day Mean","Upper","Lower"]].tail(200).plot(figsize = (16,8))
```

### Basic Analysis

```python
amz["Returns"] = ((amz["Close"]/amz["Close"].shift(1))-1) * 100
```

```python
amz["Returns"]
```

```python
googl["Returns"] =  googl["Close"].pct_change(1) * 100
```

```python
googl["Returns"]
```

```python
msft["Returns"] =  msft["Close"].pct_change(1) * 100
```

```python
f,axes = plt.subplots(3,1,figsize=(12,8),sharex=True,sharey=True)
amz["Returns"].plot(label="Amazon",title = "Returns",ax = axes[0],color = "r")
googl["Returns"].plot(label="Google",ax = axes[1],color = "g")
msft["Returns"].plot(label="Microsoft",ax = axes[2])
plt.legend()
plt.show()
```

```python
amz.index[amz["Returns"].argmax()]
```
